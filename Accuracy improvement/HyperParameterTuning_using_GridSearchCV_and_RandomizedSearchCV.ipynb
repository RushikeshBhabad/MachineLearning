```python
# Importing the dependencies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# Loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# Print the first 5 rows of the dataframe
data_frame.head()
```

```python
# Adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# Print the first 5 rows of the dataframe
data_frame.head()
```

```python
# Number of rows and Columns in this dataset
data_frame.shape
```

```python
# Checking for missing values
data_frame.isnull().sum()
```

```python
# Checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
 X = np.asarray(X)
 Y = np.asarray(Y)
```

##**GridSearchCV**

Grid Search Cross-Validation (GridSearchCV) is like a matchmaker for your model and hyperparameters. It tries out every possible combination of your hyperparameters and picks the best one based on performance

```python
# Loading the SVC model
model = SVC()
```

```python
# Hyperparameters
parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1, 5, 10, 20]
}
```

```python
# Grid Search
grid_search = GridSearchCV(model, parameters, cv=5)
```

```python
# Fitting the data to our model
grid_search.fit(X,Y)
```

```python
grid_search.cv_results_
```

```python
# Best parameters
best_parameters = grid_search.best_params_
print(best_parameters)
```

```python
# Highest accuracy
best_result = grid_search.best_score_
print(best_result)
# Accuracy by {'C': 10, 'kernel': 'linear'}
```

```python
# Loading the result to pandas dataframe
result = pd.DataFrame(grid_search.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
print(grid_search_result)
```

Highest Accuracy : 95.2%

Best Parameters : {'C': 10, 'kernel': 'linear'}


##**RandomizedSearchCV**

RandomizedSearchCV randomly samples from a range of hyperparameters instead of trying every single combination like Grid Search. This saves a ton of time ⏳ especially when you’ve got a big parameter space.

```python
# Loading the SVC model
model = SVC()
```

```python
# Hyperparameters
parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1, 5, 10, 20]
}
```

```python
# Randomized Search
random_search = RandomizedSearchCV(model, parameters, cv=5)
```

```python
# Fitting the data to our model
random_search.fit(X,Y)
```

```python
random_search.cv_results_
```

```python
# Best parameters
best_parameters = random_search.best_params_
print(best_parameters)
```

```python
# Highest accuracy
best_result = random_search.best_score_
print(best_result)
# Accuracy by {'C': 10, 'kernel': 'linear'}
```

```python
# Loading the result to pandas dataframe
result = pd.DataFrame(random_search.cv_results_)
```

```python
result.head()
```

```python
random_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
print(random_search_result)
```

Highest Accuracy : 95.08%

Best Parameters : {'kernel': 'linear', 'C': 5}
