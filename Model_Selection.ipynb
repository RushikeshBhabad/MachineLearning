```python
# importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
# importing the models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# loading the csv data to a Pandas DataFrame
dataset = pd.read_csv('/content/heart.csv')
```

```python
# print first 5 rows of the dataset
dataset.head()
```

```python
# checking the distribution of Target Variable
dataset['target'].value_counts()
```

```python
# Splitting the Features and Target
X = dataset.drop(columns='target', axis=1)
Y = dataset['target']
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

 **Model Selection**

 Comparing the models with default hyperparameter values using Cross Validation

```python
# list of models
models = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier(random_state=0)]
```

```python
def compare_models_by_cross_validation():

  for model in models:

    cv_score = cross_val_score(model, X, Y, cv=5)
    mean_accuracy = sum(cv_score)/len(cv_score)
    mean_accuracy = mean_accuracy*100
    mean_accuracy = round(mean_accuracy, 2)

    print('Cross Validation accuracies for the',model,'=', cv_score)
    print('Acccuracy score of the ',model,'=',mean_accuracy,'%')
    print('---------------------------------------------------------------')
```

```python
compare_models_by_cross_validation()
```

Comparing the models with different Hyperparameter values using GridSearchCV


```python
# list of models
models_list = [LogisticRegression(max_iter=10000), SVC(), KNeighborsClassifier(), RandomForestClassifier(random_state=0)]
```

```python
# creating a dictionary that contains hyperparameter values for the above mentioned models
model_hyperparameters = {

    'Logistic_hyperparams' : {
        'C' : [1,5,10,20]
    },
    'SVC_hyperparams' : {
        'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
        'C' : [1,5,10,20]
    },
    'KNN_hyperparameters':{
        'n_neighbors' : [3,5,10]
    },
    'Random_forest_hyperparameters' : {
        'n_estimators' : [10, 20, 50, 100]
    }

}
```

```python
type(model_hyperparameters)
print(model_hyperparameters.keys())
```

```python
model_keys = list(model_hyperparameters.keys())
print(model_keys)
```

```python
# Applying GridSearchCV
def model_selection_by_gridSearch(models_list,model_hyperparameters):
  result = []
  i = 0

  for model in models_list:
    key = model_keys[i]
    params = model_hyperparameters[key]
    i += 1

    print('Model Name : ', model)
    print('Parameters : ',params)
    print('---------------------------------')

    grid_search = GridSearchCV(model, params, cv=5)

    # fitting the data
    grid_search.fit(X,Y)
    result.append({
        'model used' : model,
        'highest score' : grid_search.best_score_,
        'best hyperparameters' : grid_search.best_params_
    })

    result_dataframe = pd.DataFrame(result, columns = ['model used','highest score','best hyperparameters'])
  return result_dataframe
```

```python
model_selection_by_gridSearch(models_list,model_hyperparameters)
```
