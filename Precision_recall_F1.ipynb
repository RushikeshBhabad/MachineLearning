True Positive (TP):
✅ Actual = Yes, Predicted = Yes
(Model was right — said yes, and it was actually yes)

False Positive (FP):
❌ Actual = No, Predicted = Yes
(Model was wrong — said yes, but it was actually no)

False Negative (FN):
❌ Actual = Yes, Predicted = No
(Model was wrong — said no, but it was actually yes)

True Negative (TN):
✅ Actual = No, Predicted = No
(Model was right — said no, and it was actually no)

1 st -> Models Output

2 nd -> Actual Output
